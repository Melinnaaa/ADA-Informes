# **1. Problema de ordenamiento**

**Entrada:** Arreglo o lista de *n* elementos comparables.

**Salida:** Permutación ordenada de la secuencia de entrada, de tal manera que el arreglo original estará ordenado de menor a mayor (*b<sup>'</sup><sub>1</sub> =< b<sup>'</sup><sub>2</sub> =< ... =< b<sup>'</sup><sub>n</sub>*).

El tamaño del problema es indicado por la cantidad de datos ingresados en la entrada, por lo cual correspondería a **n**.

El problema de ordenamiento siempre ha sido un tema del que hablar, y a este se le han buscado distintas soluciones, tales como: *QuickSort, BubbleSort, MergeSort, entre otros*. En este caso toca hablar del InsertionSort, el cual es un ordenamiento muy común para el ser humano y este puede ser representado al ordenar una mano de cartas, en donde vamos añadiendo las cartas de derecha a izquierda, para luego dejarlas compararlas con el resto y encontrar la posición correcta de esta.

# **2. InsertionSort**

## 2.1 Código

El siguiente código muestra una implementación del algoritmo **InsertionSort.**

```python
import random
from termcolor import colored
import copy

def insertion_Sort(array, verbose = False):
    n = len(array) # Largo del arreglo
    cont = 0 # Contador de comparaciones
    tmp = 0 # Contador de retrocesos por pasada
    already_sorted = False # Bandera que indica si el arreglo esta ordenado
    if (verbose == True): print("input array:",array)

    # Se recorre el arreglo, partiendo desde 1
    for i in range(1, n):
        tmp = 0
        if (verbose == True): print("\nPasada", i, "\n")
        
        pos = array[i] # Valor del elemento en la posición i
        j = i-1 #Indica el valor anterior a la posición i

        # Mientras la posición j sea mayor o igual a 0 y este sea mayor al
        # elemento i-esimo, se ordenarán.
        while (j >= 0 and pos < array[j]):
          cont +=1
          tmp = tmp + 1
          if (verbose == True): 
            print(f"Retroceso numero {tmp}")
            print(f"Se intercambia {array[j]} por {pos}")
          
          # Se realiza el swap
          array[j+1] = array[j]
          j -= 1
          array[j+1] = pos

          if (verbose == True): 
            print(f"{str(array[:j+1])[1:-1]}", ",", colored(f"{str(array[j+1:j+3])[1:-1]}","red"), ",", f"{str(array[j+3:])[1:-1]}")
            print("\n")

        if (already_sorted == False and verbose == True): 
           if (tmp == 0): print(f"No se realizaron retrocesos en la pasada.\n")
           print("Al finalizar pasada:")
           print(colored(f"{str(array[:i])[1:-1]}","blue"), ",", str(array[i:])[1:-1])

    already_sorted = True
    if (already_sorted == True):
        if (verbose == True): print("is sorted!")
    if (verbose == True): print("\noutput array:",array)

    return array, cont

# Ejemplo
A = [5,8,7,1,3]
print("Entrada: ",A)
A, counter = insertion_Sort(A)
print("Salida: ",A)
print("# comparaciones: ", counter)
print("Media de retrocesos: ", counter/len(A))
```

## 2.2. Descripción del algoritmo

El algoritmo recibe como entrada una lista (o arreglo) ***array*** con una cantidad de **n** elementos a ordenar. Luego, a través de iteraciones los números se ordenan dentro de la misma lista.


1.   En cada iteración *i*, el algoritmo avanza dentro del arreglo (de izquierda a derecha), en donde compara cada elemento de la posición `array[i]` con el elemento anterior `array[j]`, en donde j es i - 1. Si `array[i] < array[j]`, los elementos intercambian posiciones y el j se disminuye. Para que de esta manera los elementos estén ordenados de manera ascendente al finalizar cada iteración.

2.   El algoritmo finaliza al recorrer todo el arreglo. Si los valores a comparar deben intercambiarse se vuelve a realizar el paso 1, de lo contrario el algoritmo sigue recorriendo el arreglo.

3.   Al finalizar el programa este retorna el arreglo ordenado y la cantidad de comparaciones realizadas `T`.


Si la variable `verbose` es `True` se muestra información para ver lo que ocurre paso a paso dentro de la función.






## 2.3. Ejemplo


Consideremos, por ejemplo, el arreglo:

*a = [4,3,2,1]*

En la primera pasada intercambiamos el 3 por el 4:

*a = [3,4,2,1]*

Luego, en la segunda pasada el 2 se intercambia con el 4 y el 2 por el 3:

*a = [2,3,4,1]*

Y en la tercera pasada intercambia el 1 con el 4, luego el 1 con el 3 y el 1 con el 2:

*a = [1,2,3,4]*

Al finalizar de recorrer el arreglo este se encontraría ordenado.

#**3. Tiempo de ejecución**

## **Teorema (Tiempo de ejecución).**

El algoritmo **InsertionSort** tiene un **tiempo de ejecución de** $O(n^2)$ en el peor caso.

### *Prueba del teorema*

Observando el código, podemos apreciar que el **peor caso** ocurre cuando el arreglo se encuentra ordenado de mayor a menor. En este caso, en cada pasada el
elemento se movería una casilla a la izquierda y se verificaría que este no sea menor al anterior, hasta encontrar la posición correcta.

```python
A = [6,5,4,3,2,1];
print("Entrada no ordenada: ", A)
A, counter = insertion_Sort(A, True)
print("Salida ordenada: ",A)
print("Total de comparaciones realizadas: ", counter)
```

En este caso, la primera pasada hizo 1 comparación, la segunda 2, en la tercera 3 y asi sucesivamente. Con un total de 15 comparaciones.

Utilizando la formula $\sum\limits_{i=0}^{n-1} i=\frac{n(n-1)}{2}$ llegamos a la misma cantidad de comparaciones, lo cual corrobora que el tiempo de ejecución en el peor caso es $O(n^2)$.

# **4. Correctitud**

##**Teorema (Correctitud).**

El algoritmo **InsertionSort** genera un arreglo: `[a1,a2,...an]`, con los mismos elementos del arreglo de entrada, pero ordenados de menor a mayor, es decir `a1<=a2<=...<=an`.











###**Prueba del Teorema**

**Inicialización**
 
Al comienzo de la primera iteración (i=1), el elemento ubicado en la primera posición (i-1) `array[0]` se encuentra ordenado.

**Mantención**

En la posición *i-ésima* se asume que todos los elementos del arreglo hasta la posicón *i-1* se encuentran ordenados y contienen los elementos del arreglo original. 

Al momento de ejecutarse la iteración *i-ésima*, realizamos un "swap" al elemento ubicado en la posición *i*, hasta encontrar la posición correcta de este en el arreglo. Al realizar la iteración, el arreglo contendrá los elementos del arreglo original y además estos se encontrarán ordenados.

A continuación se muestra un ejemplo que comprueba la correctitud del *bucle invariante*.

```python
A = [2,6,1,4]
print("Entrada: ",A)
A = insertion_Sort(A, verbose=True)
```

Como se puede observar, al inicio de cada iteración todos los elementos hasta la posición i-1 se encuentran ordenados. Al finalizar el bucle, se genera un arreglo con los mismos elementos de entrada, pero ordenados de menor a mayor, de esta manera cumpliendo la propiedad antes dicha.

## 5.2. ¿Que ocurre al utilizar un arreglo casi completamente ordenado?

En el siguiente gráfico se muestra la cantidad de operaciones (swaps) que realizan los arreglos casi completamente ordenados, para poder lograr esto los arreglos son ordenados mediante la función *InsertionSort* y luego 3 posiciones de estos son desordenados aleatoriamente. Estos arreglos tendran un largo entre $5$ y $50$. Además, se compararán la cantidad de comparaciones realizadas con un algoritmo generado de manera aleatoria.

```python
import matplotlib.pyplot as plt 
import numpy as np
x=[n for n in range(5,50)] 
y=[]; 
x1 = [n for n in range(5,50)]
y2 = [];

for n in range(5,50):
  a = random.sample(range(1, 100), n)
  #Se ordena el arreglo 
  a, counter = insertion_Sort(a)
  y2.append(counter)

  #Se desordenan 3 posiciones aleatorias del arreglo
  for i in range (0, 3):
    rand = random.randint(0, n - 1)
    rand2 = random.randint(0, n - 1)
    #Se realiza el intercambio.
    a[rand], a [rand2] = a[rand2], a[rand] 
  a,counter = insertion_Sort(a)
  y.append(counter)

plt.plot(x,y)
plt.plot(x1,y2)
plt.legend(["InsertionSort casi ordenado", "InsertionSort random"])

plt.xlabel('Largo del arreglo')
plt.ylabel('Cantidad de operaciones')
plt.show()
```

### Análisis de resultados

Al observar el gráfico, podemos notar que la cantidad de intercambios (restrocesos) es muy irregular, ya que a pesar de que el arreglo se encuentre casi en su totalidad ordenado, la cantidad de comparaciones depende de lo alejado que esta el dato de su posición correcta. Por otro lado, la cantidad de operaciones disminuye bastante en comparación a la de un arreglo generado de manera aleatoria.

## 5.3. Tiempo de ejecución (InsertionSort vs BubbleSort)

En los siguientes gráficos se reporta el tiempo de ejecución de los algoritmos de ordenamiento **InsertionSort** y **BubbleSort** en función del tamaño del problema para arreglos generados aleatoriamente de tamaños entre $5$ y $500$. A partir de los gráficos obtenidos se realizará una comparación del rendimiento de ambos algoritmos.

### Tiempo de ejecución *InsertionSort*

```python
import matplotlib.pyplot as plt
import datetime
from timeit import repeat

x=[]; y=[]

for n in range(5,500):

  a = random.sample(range(1, 1000), n)

  #la función repeat está ejecutando 20 veces insertion_Sort con el arreglo a y retorna el mejor tiepo obtenido.
  t = repeat(setup="from __main__ import insertion_Sort", stmt=f"insertion_Sort({a})", repeat=1, number=10)

  x.append(n)
  y.append(t)


plt.plot(x,y)
plt.legend(["Insertion Sort"])

plt.xlabel('Tamaño del arreglo')
plt.ylabel('Tiempo en ms')
plt.show()
```

### Tiempo de ejecución *BubbleSort*

```python
import matplotlib.pyplot as plt
import datetime
from timeit import repeat

#Función que contiene el algoritmo de bubbleSort
def bubble_sort(a):
    n = len(a)
    T = 0 #contador de comparaciones

    for i in range(1,n+1):
        # Flag que indica si el arreglo ya se encuentra ordenado
        already_sorted = True

        # Se recorre el arreglo
        for j in range(n - i):
            T +=1
            if a[j] > a[j + 1]:
                # Si el elemento es mayor al siguiente se intercambia
                a[j], a[j + 1] = a[j + 1], a[j]
                already_sorted = False

        if already_sorted:
            break

    return a, T


x=[]; y=[]

for n in range(5,500):

  a = random.sample(range(1, 1000), n)

  #la función repeat está ejecutando 20 veces bubble_sort con el arreglo a y retorna el mejor tiepo obtenido.
  t = repeat(setup="from __main__ import bubble_sort", stmt=f"bubble_sort({a})", repeat=1, number=10)

  x.append(n)
  y.append(t)


plt.plot(x,y)
plt.legend(["BubbleSort"])

plt.xlabel('Tamaño del arreglo')
plt.ylabel('Tiempo en ms')
plt.show()
```

### **Análisis de resultados**



Al comparar los gráficos, se puede concluir que el algoritmo de busqueda *InsertionSort* es mas eficiente en arreglos mas pequeños, superando en ms en todo momento al BubbleSort entre $n \in [0,200]$. 

Por otro lado, en algoritmos de un tamaño mayor tienen un comportamiento mas o menos similar, pero de igual manera el *InsertionSort* sigue siendo mas eficiente en ms, ya que nunca alcanza una cantidad mayor a los 25 ms, en cambio el *BubbleSort* supera dicho tiempo, llegando hasta casi a los 30 ms. 

Por ultimo, podemos notar que ambos algoritmos tienen *peaks*, en los cuales seguramente estuvieron mas cerca del peor caso esperado o por el comportamiento impredecible de los procesadores.
