# **1. Problema de ordenamiento**

**Entrada:** Secuencia de *n* numeros.

**Salida:** Permutación ordenada de la secuencia de entrada, de tal manera que el arreglo original estará ordenado de menos a mayor (*b<sup>'</sup><sub>1</sub> =< b<sup>'</sup><sub>2</sub> =< ... =< b<sup>'</sup><sub>n</sub>*).

## 2.2. Descripción del algoritmo

El algoritmo recibe como entrada una lista (o arreglo) ***array*** con una cantidad de **n** elementos a ordenar. Luego, a través de iteraciones los números se ordenan dentro de la misma lista.


1.   En cada iteación *i*, el algoritmo avanza dentro del arreglo (de izquierda a derecha), en donde compara cada elemento de la posición `array[i]` con el elemento anterior `array[j]`, en donde j es i - 1. Si `array[i] < array[j]`, los elementos intercambian posiciones y el j se disminuye.

2.   El algoritmo finaliza al recorrer todo el arreglo. Si los valores a comparar deben intercambiarse se vuelve a realizar el paso 1, de lo contrario el algoritmo sigue recorriendo el arreglo.

3.   Al finalizar el programa este retorna el arreglo ordenado y la cantidad de comparaciones realizadas `T`


Si la variable `verbose` es `True` se muestra información para ver lo que ocurre paso a paso dentro de la función.

# 4. Correctitud

**Teorema (Correctitud).**

El algoritmo **InsertionSort** genera un arreglo: `[a1,a2,...an]`, con los mismos elementos del arreglo de entrada, pero ordenados de menor a mayor, es decir `a1<=a2<=...<=an`

**Prueba del Teorema**

*Inicialización*
 
Al comienzo de la primera iteración (i=1), el elemento ubicado en la primera posición `array[0]` se encuentra ordenado.

*Mantención*

En la posición *i-esima* se asume que todos los elementos del arreglo hasta la posicón *i-1* se encuentran ordenados y contienen los elementos del arreglo original. 

Al momento de ejecutarse la iteración *i-esima*, realizamos un "swap" al elemento ubicado en la posición *i*, hasta encontrar la posición correcta de este en el arreglo. Al realizar la iteración, el arreglo contendrá los elementos del arreglo original y ademas estos se encontrarán ordenados.

Tiempo de ejecución *BubbleSort*

```python
import matplotlib.pyplot as plt
import datetime
from timeit import repeat

#Función que contiene el algoritmo de bubbleSort
def bubble_sort(a):
    n = len(a)
    T = 0 #contador de comparaciones

    for i in range(1,n+1):
        # Flag que indica si el arreglo ya se encuentra ordenado
        already_sorted = True

        # Se recorre el arreglo
        for j in range(n - i):
            T +=1
            if a[j] > a[j + 1]:
                # Si el elemento es mayor al siguiente se intercambia
                a[j], a[j + 1] = a[j + 1], a[j]
                already_sorted = False

        if already_sorted:
            break

    return a, T


x=[]; y=[]

for n in range(5,500):

  a = random.sample(range(1, 1000), n)

  #la función repeat está ejecutando 20 veces bubble_sort con el arreglo a y retorna el mejor tiepo obtenido.
  t = repeat(setup="from __main__ import bubble_sort", stmt=f"bubble_sort({a})", repeat=1, number=10)

  x.append(n)
  y.append(t)


plt.plot(x,y)
plt.legend(["BubbleSort"])

plt.xlabel('n')
plt.ylabel('time in ms')
plt.show()
```

Al comparar los gráficos, se puede concluir que el algoritmo de busqueda "InsertionSort" es mas eficiente tanto en arreglos mas pequeños, superando en ms en todo momento al BubbleSort.
